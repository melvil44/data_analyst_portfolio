```python
import pandas as pd

# Load your dataset
df = pd.read_csv('Production_Crops_E_Asia.csv')

# Display the first few rows of the dataset to understand its structure
df.head()
```

```python
# Filter rows where the "element code" column is equal to 5510
df_prodonly = df[df['Element Code'] == 5510]

# Display the first few rows of the filtered DataFrame
df_prodonly.head()
```

```python
# Filter out columns ending with "F"
df_noflag = df_prodonly[[col for col in df_prodonly.columns if not col.endswith('F')]]

# Display the first few rows of the filtered DataFrame
df_noflag.head()
```

```python
# List of columns to keep (those not in the range Y1961 to Y1979)
columns_to_keep = [col for col in df_noflag.columns if not col.startswith('Y') or int(col[1:]) < 1961 or int(col[1:]) > 1979]

# Create a new DataFrame with only the columns to keep
df_lessyears = df_noflag[columns_to_keep]

# Display the first few rows of the filtered DataFrame
df_lessyears.head()
```

```python
# Define the list of items to keep
items_to_keep = ['Cereals, Total', 'Vegetables Primary', 'Fruit Primary', 'Treenuts, Total']

# Filter rows where the "Item" column is in the list of items to keep
df_gooditems = df_lessyears[df['Item'].isin(items_to_keep)]

# Display the first few rows of the filtered DataFrame
df_gooditems.head(10)
```

```python
# Identify the columns you want to keep fixed (non-pivoted)
fixed_columns = ['Area Code', 'Area', 'Item Code', 'Item', 'Element Code', 'Element', 'Unit']

# Pivot the DataFrame
df_pivoted = pd.melt(df_gooditems, id_vars=fixed_columns, var_name='Year', value_name='Production')

# Display the first few rows of the pivoted DataFrame
df_pivoted.tail(100)
```

```python
# Sort the DataFrame by "Item" and "Year" columns
df_sorted = df_pivoted.sort_values(by=['Area', 'Item', 'Year'])
df_sorted.tail(100)
```

```python
df_sorted.head(100)
```

```python
# Save the sorted DataFrame to a new CSV file
df_sorted.to_csv('sorted_data.csv', index=False)
```

```python
# Assuming your DataFrame is named df
# Remove the "Y" from all the cells in the "year" column
df_sorted['Year'] = df_sorted['Year'].str.replace('Y', '')

# Convert the "year" column to integer data type
df_sorted['Year'] = df_sorted['Year'].astype(int)

# Verify the changes
print(df_sorted['Year'].head())
```
